```python
import os
import pandas as pd
from sqlalchemy import create_engine, text

# Definindo os parâmetros de conexão para o MySQL
user = 'root'
password = 'root'
host = 'localhost'

# Definindo o caminho base onde estão os diretórios
base_dir = '../data/raw'

# Função para carregar arquivos CSV para o banco de dados
def load_csv_to_db(engine, csv_path, table_name):
    # Carregando o arquivo CSV em um DataFrame
    df = pd.read_csv(csv_path)
    
    # Carregando o DataFrame para o banco de dados
    df.to_sql(table_name, con=engine, if_exists='replace', index=False)
    print(f"Tabela {table_name} criada no banco de dados.")

# Conectando ao banco de dados 'mysql' para criar novos bancos de dados
engine_root = create_engine(f'mysql+pymysql://{user}:{password}@{host}/mysql')

# Iterando sobre os diretórios dentro de 'data/row'
for subdir in os.listdir(base_dir):
    subdir_path = os.path.join(base_dir, subdir)

    # Verifica se o subdiretório é realmente um diretório
    if os.path.isdir(subdir_path):
        # Criando o banco de dados com o nome do diretório
        database_name = subdir
        
        # Criando o banco de dados (se não existir)
        with engine_root.connect() as connection:
            connection.execute(text(f"CREATE DATABASE IF NOT EXISTS `{database_name}`;"))
            print(f"Banco de dados {database_name} criado.")
        
        # Criando uma nova engine para o banco de dados recém-criado
        db_url = f'mysql+pymysql://{user}:{password}@{host}/{database_name}'
        engine = create_engine(db_url)

        # Iterando sobre os arquivos dentro do subdiretório
        for root, dirs, files in os.walk(subdir_path):
            for file in files:
                if file.endswith('.csv'):
                    # Pegando o caminho completo do arquivo
                    csv_path = os.path.join(root, file)
                    
                    # O nome da tabela será o nome do arquivo sem a extensão
                    table_name = os.path.splitext(file)[0]
                    
                    # Carregando o CSV no banco de dados
                    load_csv_to_db(engine, csv_path, table_name)
```
